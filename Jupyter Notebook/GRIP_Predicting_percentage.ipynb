# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
dataframe = pd.read_csv("GRIP Study Hours.csv")
```

```python
dataframe.head()
```

```python
plt.scatter(dataframe['Hours'], dataframe['Scores'], color = 'blue')
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
dataframe.shape
```

There are 25 rows and 2 columns in dataset

# Checking for missing values(if any present)

```python
dataframe.isnull().sum().any()
```

```python
x = dataframe.iloc[:, :-1].values
y = dataframe.iloc[:, -1].values
```

# Splitting the dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .10)
```

# Training

```python
from sklearn.linear_model import LinearRegression
```

```python
Lin_Reg = LinearRegression()
Lin_Reg.fit(x_train, y_train)
```

```python
y_pred = Lin_Reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```

# Visualisation


```python
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, Lin_Reg.predict(x_train), color = 'red')
plt.title('Hours vs Scores (Training set)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

# Predicting Score

```python
prediction = Lin_Reg.predict([[9.25]])
prediction
```

The student will score around 92.68% if he/she will study for 9.25 hours per day.
